#### Read the dataset assets/real_estate.csv and try to visualize the table (★☆☆)

```python
# Import Py Packages for Data Management
import pandas as pd
import matplotlib.pyplot as plt

# Import Real Estate Dataaset
file_path = 'assets/real_estate.csv'
df = pd.read_csv(file_path, delimiter=';', header=0, encoding='utf-8')

# Visualize first 10 lines and get full dataset info
df.head(10)
```

#### Now let's work with the "south belt" of madrid. Make a subsetting that contains the following populations (level5): "Fuenlabrada","Leganés","Getafe","Alcorcón" (★★☆) 

```python
# South Belt of Madrid Data

# List of values to subset
populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Create a boolean mask for filtering
mask = df['level5'].isin(populations)

# Apply the mask to the DataFrame to create the subset
subset_df = df[mask]
```

#### Make a bar plot of the median of prices (★★★) 

```python
# Creat a bar plot for Median of Prices of Madrid South Belt


# Calculate the median price for each location
median_prices = subset_df.groupby('level5')['price'].median()

# Create a bar plot of the median prices with differences
fig, ax = plt.subplots(figsize=(6, 5))
median_prices.plot.bar(rot=0, ax=ax)

# Set plot title and labels
plt.title("Median House Price in South Belt")
plt.xlabel('Location')
plt.ylabel('Median Price')
for i, v in enumerate(median_prices):
    ax.text(i, v, str(v), ha='center', va='bottom')

# Display Bar Plot
plt.show()
```

#### Calculate the sample mean and variance of the variables: price, rooms, surface area and bathrooms (★★★) 

```python
# Calculate the sample mean for each variable
mean_price = subset_df['price'].mean()
mean_rooms = subset_df['rooms'].mean()
mean_surface_area = subset_df['surface'].mean()
mean_bathrooms = subset_df['bathrooms'].mean()

# Calculate the sample variance for each variable
var_price = subset_df['price'].var()
var_rooms = subset_df['rooms'].var()
var_surface_area = subset_df['surface'].var()
var_bathrooms = subset_df['bathrooms'].var()

# Print the results
print("Sample Mean:")
print(f"Price: {round(mean_price)}")
print(f"Rooms: {round(mean_rooms)}")
print(f"Surface Area: {round(mean_surface_area)}")
print(f"Bathrooms: {round(mean_bathrooms)}")

print("\nSample Variance:")
print(f"Price: {var_price}")
print(f"Rooms: {var_rooms}")
print(f"Surface Area: {var_surface_area}")
print(f"Bathrooms: {var_bathrooms}")
```

#### Normalize the variable of prices for each population and plot the 4 histograms in the same plot (★★★) 

```python
# Normalize the prices for each population

# List of values to subset
populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Create a boolean mask for filtering
mask = df['level5'].isin(populations)

# Apply the mask to the DataFrame to create the subset
subset_df = df[mask].copy()

# Normalize the 'price' variable 
#subset_df['price_normalized'] = (subset_df['price'] - subset_df['price'].mean()) / subset_df['price'].std() # z-score norm is also good
subset_df['price_normalized'] = (subset_df['price'] - subset_df['price'].min()) / (subset_df['price'].max() - subset_df['price'].min())

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(13, 8))
axs = axs.flatten() 

# Loop through each population and create histograms in the subplots
for i, population in enumerate(populations):
    subset_population = subset_df[subset_df['level5'] == population]
    axs[i].hist(subset_population['price_normalized'], bins=40, alpha=0.5)
    axs[i].set_title(population)
    axs[i].set_xlabel('Normalized Price')
    axs[i].set_ylabel('Frequency')

# Add overall title
fig.suptitle('Histograms for South Belt of Madrid Price (Normalized)')

# Adjust spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

#### Make a plot of the coordinates (lat and long) of the south belt of Madrid by color of each population (★★★★) 
This is a extremely hard exercise and we highly recommend to do it at the end of everything

```python
# Import necessary libraries
from ipyleaflet import Map, basemaps, Marker, AwesomeIcon

# Create a map object
map = Map(center=(60, -2.2), zoom=2, min_zoom=1, max_zoom=20, basemap=basemaps.Stamen.Terrain)

# List of values to subset
populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Define a dictionary to map values to marker colors
color_dict = {"Fuenlabrada": "blue", "Leganés": "green", "Getafe": "red", "Alcorcón": "purple"}

# Create a boolean mask for filtering
mask = df['level5'].isin(populations)

# Apply the mask to the DataFrame to create the subset
subset_df = df[mask]

# Filter out rows with 0 latitude and 0 longitude
subset_df = subset_df[(subset_df['latitude'] != '0') & (subset_df['longitude'] != '0')]

# Use list comprehension to create tuples from latitude and longitude columns
lat_list = [float(lat.replace(',', '.')) for lat in subset_df['latitude']]
long_list = [float(long.replace(',', '.')) for long in subset_df['longitude']]


# Loop through the latitude and longitude lists and add markers to the map
for i, (lat, long) in enumerate(zip(lat_list, long_list)):
    population = subset_df.iloc[i]['level5']
    color = color_dict.get(population, "black") 
    marker = Marker(location=(lat, long), icon=AwesomeIcon(name='house', marker_color=color), draggable=False)
    map.add_layer(marker)

# Display the map
map

## Stuff I would change to make more sense:
# Create an ipyleaflet map centered on the points
#map = Map(center=(40.4168, -3.7038), zoom=10) # Set the center and zoom level of the map

# Glad to have not used iterrows as it's slower to shrug big data
# Initally had issue because lat and long where objects and not floats. Converted them. 
# I feel the code can be simpler but I don't know how. 
# I can try to vectorize some code with pd as I've read it's more efficient. 

```

#### What can you say about the price per square meter between the towns of Getafe and Alcorcón?

```python
# List of values to subset
populations2 = ["Getafe", "Alcorcón"]

# Create a boolean mask for filtering
mask2 = df['level5'].isin(populations2)


# Apply the mask to the DataFrame to create the subset
subset_df2 = df[mask2].copy()  # Make a copy to avoid the SettingWithCopyWarning

# Calculate price per square meter
subset_df2['price_per_square_meter'] = subset_df2['price'] / subset_df2['surface']

# Plot the price per square meter for the subset
fig, ax = plt.subplots()
subset_df2.groupby('level5')['price_per_square_meter'].plot(kind='hist', ax=ax, alpha=0.5, bins=30)
ax.legend(populations2)
ax.set_xlabel('Price per Square Meter')
ax.set_ylabel('Frequency')
ax.set_title('Comparison of Price per Square Meter for Getafe and Alcorcón')
plt.show()


# We can say that Getafe prices are very stable onto mid-range 1500-2500 eur, which
# makes it a a mid-level zone. 
# Alcorcón has a vaster offer, main price is onto 1500-1700 eur per square
# meter, but also has more expensive houses than Getafe. They are similar areas
# but there are cheaper houses in Alcorcón vs Getafe and also more expensive. 
```

#### Reload the data, is the average of "Valdemorillo" and "Galapagar" prices the same?

```python
# List of values to subset
valde = ["Valdemorillo"]
galap = ["Galapagar"]

# Create a boolean mask for filtering
maskval = df['level5'].isin(valde)
maskgal = df['level5'].isin(galap)

# Apply the mask to the DataFrame to create the subset
subset_valde = df[maskval]
subset_galap = df[maskgal]

# Calculate the median price for each location
avg_price_val = subset_valde.groupby('level5')['price'].mean()
avg_price_gal = subset_galap.groupby('level5')['price'].mean()

# Plot the avg price 
plt.bar(avg_price_val.index, avg_price_val.values, label='Valdemorillo')
plt.bar(avg_price_gal.index, avg_price_gal.values, label='Galapagar')
plt.xlabel('Location')
plt.ylabel('Average Price')
plt.title('Comparison of Average Price by Location')
plt.legend()
plt.show()

print(avg_price_val)
print(avg_price_gal)


# Conclusion: Galapagar is cheaper in average to Valdemorillo by 3797 eur


## Upon researching this is much simpler way to subset the data and write code: 
#avg_price_valde = df.query("level5 == 'Valdemorillo'")['price'].mean()
#avg_price_galap = df.query("level5 == 'Galapagar'")['price'].mean()

# Plot the avg price 
#plt.bar(['Valdemorillo', 'Galapagar'], [avg_price_valde, avg_price_galap])

```

#### Could you say there is a relation between surface and price? (★★☆) 
Hint: make a scatter plot.

```python
# List of values to subset
populations3 = ["Valdemorillo", "Galapagar"]

# Create a boolean mask for filtering
mask3 = df['level5'].isin(populations3)


# Apply the mask to the DataFrame to create the subset
subset_df3 = df[mask3].copy()  # Make a copy to avoid the SettingWithCopyWarning

# Calculate price per square meter
subset_df3['price_per_square_meter'] = subset_df3['price'] / subset_df3['surface']

# Create a scatter plot for price vs surface
fig, ax = plt.subplots()
for population in populations3:
    subset_population = subset_df3[subset_df3['level5'] == population]
    ax.scatter(subset_population['surface'], subset_population['price'], alpha=0.5, label=population)

ax.legend()
ax.set_xlabel('Surface')
ax.set_ylabel('Price')
ax.set_title('Scatter Plot of price/m2 for Valdemorillo and Galapagar')
plt.show()

# Yes there is a relation, the bigger the surface the more expensive. Despite
# having a few outliers in the data. Can be specific houses that differ due to
# luxury accomodations for example or just people asking whatever they want
# because it's a free market, and doesn't mean it's the value.
```

#### Make the same plot for 4 different populations (level5) and rearrange them on the same graph? (★★☆)

Check if there is a relationship between the number of rooms and bathrooms and the price of the houses?

```python
# Clean the NA from Bathrooms, Rooms and Price
df_clean = df.dropna(subset=['bathrooms', 'rooms'])

# Filter the Top 5 Locations with more houses to study the relation
top5 = df_clean['level5'].value_counts().head(5).index.to_list()
print(top5)

# Filter just the Top 5 Locations
filter_df = df_clean[df_clean['level5'].isin(top5)]

# Check numerical correlation between bathroom, room, price
correlation_matrix = filter_df[['bathrooms', 'rooms', 'price']].corr()
print(correlation_matrix)

# Plot a Scatter to visualize the data
plt.scatter(filter_df['bathrooms'], filter_df['rooms'], c=filter_df['price'], cmap='viridis')
plt.xlabel('Bathrooms')
plt.ylabel('Rooms')
plt.title('Relationship between Bathrooms, Rooms, and Price')
plt.colorbar(label='Price')
plt.show()

```
